## Data Cleaning (continued)

From part 3 we imputed and deleted some missing data. Now we'll continue to make some more formatting adjustments and tidy up othe areas of the dataset

```python
import pandas as  pd
import os

pd.set_option('display.max_columns', 30)

parent_path = os.path.dirname(os.path.dirname(os.getcwd()))

replace_double_slash = parent_path.replace('\\', '/')

data_path = replace_double_slash + '/data/shot_logs_clean_3.csv'

nba_step_final = pd.read_csv(data_path)
```

```python
nba_step_final.head()
```

We're going to make some formatting changes to the **Player Name** and the **Closest Defender** column. Notice that they are formatted very differently. In this step we'll be formatting the player name column to match the other column

```python
def get_ordered_names(s):
    if len(s) == 1:
        return str.title((s[0]))
    else:
        return str.title((' '.join(list(reversed(s[-1:-len(s):-1]))) + ', ' + s[0]))

Player_Names = nba_step_final['Player Name'].str.split(' ')

corrected = Player_Names.apply(get_ordered_names)

nba_step_final['Player Name'] = corrected

nba_step_final.head()
```

And that's it! Next step we'll looking into the data a little bit more and analyze patterns, visualize some player data, etc.

```python
nba_clean_final = nba_step_final

save_path = replace_double_slash + '/data/shot_logs_clean_final.csv'

nba_clean_final.to_csv(save_path, index=False)
```
